```python
from konlpy.tag import Mecab

# Mecab 초기화
mecab = Mecab(dicpath='/usr/local/lib/mecab/dic/mecab-ko-dic')

# 테스트 텍스트
text = 'SUPEX를 통한 AI리더십과 DDR5 메모리 기술'

# 형태소 분석
result = mecab.morphs(text)
print("형태소 분석 결과:", result)

# 품사 태그와 함께 출력
pos_result = mecab.pos(text)
print("\n품사 태그 분석 결과:", pos_result)
```

```python
import numpy as np
import matplotlib.pyplot as plt
from wordcloud import WordCloud
import pandas as pd
from PIL import Image

# 리더십 데이터 정의
leadership_data = {
    '탐색적 리더십(Exploration)': [
        '탐색', 'Exploration', 'CEO', '사장', '대표이사', '회장', '경영진', '임원', '수석', '총괄', 
        '책임자', '이사', '혁신', '변화', '연구', '투자', '실패 수용', '장기적 이익', '대담한 전략', 
        '혁신적 사고', '조직 변화', '비전', '권한 위임', '적응력', '유연성', '위기 관리', '변혁적 리더십', 
        '개발', '진정성 리더십', '지속가능성'
    ],
    '활용적 리더십(Exploitation)': [
        '활용', 'Exploitation', 'CEO', '사장', '대표이사', '회장', '경영진', '임원', '수석', '총괄', 
        '책임자', '이사', '성과', '성장', '경영', '전략', '위험 회피', '단기적 이익', '효율적 운영', 
        '지속적 개선', '인재 개발', '의사결정', '동기 부여', '팀워크', '소통', '성과 관리', '문제 해결', 
        '조직 문화', '고객 중심', '관계 구축', '협업', '영향력', '윤리적 리더십', '섬김의 리더십'
    ]
}

# 원형 마스크 생성
def create_circular_mask(width, height, center=None, radius=None):
    if center is None:
        center = (int(width/2), int(height/2))
    if radius is None:
        radius = min(width, height) / 2
    
    Y, X = np.ogrid[:height, :width]
    dist_from_center = np.sqrt((X - center[0])**2 + (Y - center[1])**2)
    
    mask = dist_from_center <= radius
    return 255 * mask.astype(int)

# 빈도수 계산
def calculate_frequencies(words_list):
    freq_dict = {}
    for word in words_list:
        if word in freq_dict:
            freq_dict[word] += 1
        else:
            freq_dict[word] = 1
    return freq_dict

# 원형 워드클라우드 생성 함수
def generate_circular_wordcloud(words_dict, title, color_map, mask=None):
    wordcloud = WordCloud(
        width=800, 
        height=800, 
        background_color='white',
        colormap=color_map,
        mask=mask,
        font_path='/usr/share/fonts/truetype/nanum/NanumGothic.ttf',
        min_font_size=10,
        max_font_size=120,
        prefer_horizontal=0.7,
        contour_width=1,
        contour_color='steelblue'
    ).generate_from_frequencies(words_dict)
    
    plt.figure(figsize=(10, 10))
    plt.imshow(wordcloud, interpolation='bilinear')
    plt.axis('off')
    plt.title(title, fontsize=20, pad=20)
    return wordcloud

# 워드클라우드 생성을 위한 준비
width, height = 800, 800
circle_mask = create_circular_mask(width, height)

# 한글 폰트 설정
import matplotlib.font_manager as fm
plt.rcParams['font.family'] = 'NanumGothic'
plt.rcParams['axes.unicode_minus'] = False

# 탐색적 리더십 워드클라우드
exploration_words = calculate_frequencies(leadership_data['탐색적 리더십(Exploration)'])
# 가중치 부여 (중요 키워드 강조)
exploration_words['탐색'] = 15
exploration_words['Exploration'] = 15
exploration_words['혁신'] = 12
exploration_words['변화'] = 10
exploration_words['연구'] = 10
exploration_words['투자'] = 8
exploration_words['실패 수용'] = 8

# 활용적 리더십 워드클라우드
exploitation_words = calculate_frequencies(leadership_data['활용적 리더십(Exploitation)'])
# 가중치 부여 (중요 키워드 강조)
exploitation_words['활용'] = 15
exploitation_words['Exploitation'] = 15
exploitation_words['성과'] = 12
exploitation_words['성장'] = 10
exploitation_words['경영'] = 10
exploitation_words['전략'] = 8
exploitation_words['효율적 운영'] = 8

# 워드클라우드 생성 및 저장
plt.figure(figsize=(16, 8))

plt.subplot(1, 2, 1)
exploration_cloud = generate_circular_wordcloud(
    exploration_words, 
    '탐색적 리더십 (Exploration)', 
    'viridis', 
    circle_mask
)
plt.subplot(1, 2, 2)
exploitation_cloud = generate_circular_wordcloud(
    exploitation_words, 
    '활용적 리더십 (Exploitation)', 
    'plasma', 
    circle_mask
)

plt.suptitle('양손잡이 리더십 (Ambidextrous Leadership) 워드클라우드', fontsize=22, y=0.98)
plt.tight_layout(pad=3)
plt.savefig('leadership_wordcloud.png', dpi=300, bbox_inches='tight')
plt.show()

# 결합된 워드클라우드 (원형 모양으로 구분하여 하나의 이미지로)
plt.figure(figsize=(12, 12))

# Yin-Yang 스타일의 결합 워드클라우드 만들기
combined_mask = np.zeros((width, height), dtype=np.int32)
# 왼쪽 반원 (탐색)
for i in range(width):
    for j in range(height):
        if circle_mask[j, i] > 0 and i < width/2:
            combined_mask[j, i] = 1
            
# 오른쪽 반원 (활용)
for i in range(width):
    for j in range(height):
        if circle_mask[j, i] > 0 and i >= width/2:
            combined_mask[j, i] = 2

# 각 카테고리별 색상 지정
from matplotlib.colors import ListedColormap
colors = np.array([[68, 1, 84], [253, 231, 37]])  # viridis 색상의 시작과 끝
colors = colors / 255.0  # 정규화
cmap = ListedColormap(colors)

plt.imshow(combined_mask, cmap=cmap, interpolation='bilinear')
plt.axis('off')

# 워드클라우드 타이틀
plt.title('양손잡이 리더십: 탐색과 활용의 균형', fontsize=24, pad=20)
plt.tight_layout()
plt.savefig('ambidextrous_leadership_combined.png', dpi=300, bbox_inches='tight')
plt.show()
```
